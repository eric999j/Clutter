```
def getFileName(start, end, step):
  img_paths = []
  for gif in range(start, end,step): #Check file name
    if gif<10:
      # print("frame_0"+str(gif)+".gif")
      img_paths.append("frame_0"+str(gif)+".gif")
    else:
      # print("frame_"+str(gif)+".gif")
      img_paths.append("frame_"+str(gif)+".gif")
  return img_paths
getFileName(0, 50, 5)
```

```
import imageio
def compose_gif():
    img_paths = []
    gif_images = []
    img_paths = getFileName(0, 50, 5)
    print(img_paths)
    for path in img_paths:
        gif_images.append(imageio.imread(path))
    imageio.mimsave("panda.gif",gif_images,fps=1)
compose_gif()
```
